```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed data
data = pd.read_csv('../data/processed_brent_oil_prices.csv', parse_dates=['Date'], index_col='Date')

# Plot historical prices
print("Historical Brent Oil Prices:")
print(f"1. The dataset contains {len(data)} records of daily Brent oil prices from {data.index.min().strftime('%Y-%m-%d')} to {data.index.max().strftime('%Y-%m-%d')}.")
print(f"2. The average price over the entire period is ${data['Price'].mean():.2f} per barrel.")
print(f"3. The highest price recorded was ${data['Price'].max():.2f} on {data['Price'].idxmax().strftime('%Y-%m-%d')}.")
print(f"4. The lowest price recorded was ${data['Price'].min():.2f} on {data['Price'].idxmin().strftime('%Y-%m-%d')}.")
print(f"5. The standard deviation of prices is ${data['Price'].std():.2f}, indicating significant volatility.\n")

data['Price'].plot(title='Brent Oil Prices Over Time', figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Price (USD per barrel)')
plt.grid(True)
plt.show()

# Calculate rolling mean and standard deviation
rolling_mean = data['Price'].rolling(window=30).mean()
rolling_std = data['Price'].rolling(window=30).std()

# Print rolling statistics summary
print("Rolling Statistics (30-day window):")
print(f"1. The average rolling mean over the entire period is ${rolling_mean.mean():.2f} per barrel.")
print(f"2. The highest rolling mean recorded was ${rolling_mean.max():.2f} on {rolling_mean.idxmax().strftime('%Y-%m-%d')}.")
print(f"3. The lowest rolling mean recorded was ${rolling_mean.min():.2f} on {rolling_mean.idxmin().strftime('%Y-%m-%d')}.")
print(f"4. The average rolling standard deviation is ${rolling_std.mean():.2f}, indicating the average volatility over a 30-day window.")
print(f"5. The highest rolling standard deviation recorded was ${rolling_std.max():.2f} on {rolling_std.idxmax().strftime('%Y-%m-%d')}.\n")

# Plot rolling mean and standard deviation
plt.figure(figsize=(10, 6))
plt.plot(data['Price'], label='Original')
plt.plot(rolling_mean, label='Rolling Mean (30-day)')
plt.plot(rolling_std, label='Rolling Std (30-day)')
plt.title('Brent Oil Prices with Rolling Mean and Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Price (USD per barrel)')
plt.legend()
plt.grid(True)
plt.show()
```
