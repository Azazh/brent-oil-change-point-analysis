```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('../data/BrentOilPrices.csv', parse_dates=['Date'], index_col='Date')

# Display first few rows
print("First few rows of the dataset:")
print(data.head())

# Check for missing values
print("\nMissing values in the dataset:")
print(data.isnull().sum())

# Descriptive statistics
print("\nDescriptive statistics of Brent oil prices:")
print(data['Price'].describe())

# Plot historical prices
data['Price'].plot(title='Brent Oil Prices Over Time', figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Price (USD per barrel)')
plt.grid(True)
plt.show()

# Key insights
print("\nKey Insights:")
print(f"1. The dataset contains {len(data)} records of daily Brent oil prices from {data.index.min().strftime('%Y-%m-%d')} to {data.index.max().strftime('%Y-%m-%d')}.")
print(f"2. The average price of Brent oil over the entire period is ${data['Price'].mean():.2f} per barrel.")
print(f"3. The highest price recorded was ${data['Price'].max():.2f} on {data['Price'].idxmax().strftime('%Y-%m-%d')}.")
print(f"4. The lowest price recorded was ${data['Price'].min():.2f} on {data['Price'].idxmin().strftime('%Y-%m-%d')}.")
print(f"5. The standard deviation of prices is ${data['Price'].std():.2f}, indicating significant volatility.")
```
